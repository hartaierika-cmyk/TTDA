3. Operaciones vectorizadas (15 pts)

(a) Calcula:
• a + b
• a * b

(b) Eleva cada elemento de a al cuadrado usando una funci´on vectorizada.

```python
a = np.array([1,2,3,4,5])
b = np.array([10,20,30,40,50])


sum_ab = a + b
mult_ab = a * b
div_ab = a / b

sum_ab_2 = np.add(a, b)
producto_2 = np.multiply(a, b)
division_2 = np.divide(b, a)

print(a)
print(b)

print(sum_ab)
print(mult_ab)
print(div_ab)

print(sum_ab_2)
print(producto_2)
print(division_2)

a_quad = np.square(a)
print(a_quad)
```

5. Manipulaci´on de matrices
(a) Crea una matriz M de tama˜no 4x4 con n´umeros del 1 al 16.
(b) Muestra:
• La primera fila.
• La ´ultima columna.
• La submatriz central de 2x2.
(c) Calcula la suma de cada fila y de cada columna.

```python
M = np.arange(1,17).reshape(4,4)

print(M)
print(M[0:1])
print(M[0, -1])
print(M[1:3, 1:3])

suma_filas = np.sum(M, axis = 1)
print(suma_filas)

suma_columnas = np.sum(M, axis = 0)
print(suma_columnas)
```

```python
6. Aplicaci´on pr´actica
Sup´on que tienes un array que representa las temperaturas diarias (°C) de una
ciudad durante una semana: temps = np.array([18, 20, 22, 21, 19, 17, 16])
(a) Calcula la temperatura media semanal.
(b) Muestra qu´e d´ıas estuvieron por encima del promedio.
(c) Convierte todas las temperaturas a°F usando la f´ormula:
F= C × 9/5 + 32
d) Agrega una nueva temperatura (d´ıa 8 = 23°C) al array original.
```

```python
temps = np.array([18, 20, 22, 21, 19, 17, 16])
print (temps)
temps_med = np.mean(temps)
print(temps_med)

convert_F_temps = temps * 9/5 + 32
print(convert_F_temps)

temps_new = np.append(temps, 23)
print(temps_new)

temps_new = np.concatenate((temps, [23]))
print(temps_new)
```

7. Operaciones de conjuntos
Dadas los ficheros .csv datos 1.csv y datos 2.csv (disponibles en la p´agina de la
asignatura) que contienen respectivas arrays. Calcula y muestra:
(a) El conjunto de variables estad´ısticas b´asicas:
• Media.
• Mediana.
• Valor m´aximo y m´ınimo.
• N´umero de elementos presentes en cada fichero.
(b) La intersecci´on de los elementos de la primera fila de los dos arrays.
(c) La uni´on de los elementos de las primeras filas de los dos arrays (sin duplica-
dos).
(d) Las filas repetidas en ambos ficheros.

```python
datos_1 = np.loadtxt("datos_1.csv", delimiter = ",")
datos_2 = np.loadtxt("datos_2.csv", delimiter = ",")

print(datos_1)
print("-"*53)
print(datos_2)

datos_1_mean = np.mean(datos_1)
datos_1_med = np.median(datos_1)
datos_1_max = np.max(datos_1)
datos_1_min = np.min(datos_1)
datos_1_size = datos_1.size

print(datos_1_mean)
print(datos_1_med)
print(datos_1_max)
print(datos_1_min)
print(datos_1_size)

datos_2_mean = np.mean(datos_2)
datos_2_med = np.median(datos_2)
datos_2_max = np.max(datos_2)
datos_2_min = np.min(datos_2)
datos_2_size = datos_2.size

print(datos_2_mean)
print(datos_2_med)
print(datos_2_max)
print(datos_2_min)
print(datos_2_size)

interseccion = np.intersect1d(datos_1[0], datos_2[0])
print(interseccion)

union = np.union1d(datos_1[0], datos_2[0])
print(union)

repetidas = np.array([fila for fila in datos_1 if (fila == datos_2).all(axis=1).any()])
print(np.array(repetidas))

```
