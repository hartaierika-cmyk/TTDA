```python
import pandas as pd
import numpy as np

df = pd.read_csv("WHO.csv")
print(df)
```

```python
def rename_continents(df, columna="Continent"):
    df_new = df.copy()

    # Convert to int first so 1.0, "1"  1
    df_new[columna] = df_new[columna].astype(int)

    mapping = {
        1: "Africa",
        2: "Europe",
        3: "Asia",
        4: "North America",
        5: "South America",
        6: "Oceania",
        7: "Antarctica"
    }

    df_new[columna] = df_new[columna].map(mapping)

    return df_new
```

```python
df = rename_continents(df)
df["Continent"].unique()
```

```python
df.columns
```

```python
###Categorizacin de datos
import numpy as np
import pandas as pd
np.random.seed(0)
df_random = pd.DataFrame({"candidat":np.arange(1,11), "nota":np.random.randint(0,11,size=10)})
df_random
```

```python
df_random["notaCategorica"] = pd.cut(df_random.nota, 3, labels=["dolent", "mig", "bona"])
df_random
```

```python
df_random[df_random.notaCategorica==df_random.notaCategorica.max()]
```

```python
# mejora
bins = [-1, 4, 7, 8.5, 10]
labels = ['Suspens', 'Be', 'Notable', 'Excellent']
df_random["notaCategorica"] = pd.cut(df_random.nota, bins, labels=labels) # include_lowest=True
df_random
```

```python
# Podemos definir el intervalo que deseemos (como un Histograma)
import numpy as np
marks = np.arange(0,1.1,0.25)
print(marks)
print
print("-"*10)

factors = pd.qcut(df_random.nota, marks)
print(factors)
print("-"*10)

print(pd.value_counts(factors))
print(df_random.nota)
```
