```python
df = pd.read_csv("winequality-red.csv", delimiter = ";")
df
df.describe()
```

Fixed Acidity:
Los vinos tintos tienen un nivel de 8,32 fixed acidity en promedio, el mínimo acidity es 4,6 el máximo es 15.9.
75% de los vinos tienen un nivel de fixed acidity inferior a 9,2. Desviacion estandar es bajo, 1,74 uniades en general del valor medio (8,32)

Volatile Acidity:
Los valores varían entre 0.12 y 1.58, con poca variabilidad (desviación estándar de 0.17).

Citric Acid:
El promedio de ácido cítrico es 0,27. Existen vinos con niveles de ácido cítrico igual a 0.
El 75% de los vinos tiene un nivel de ácido cítrico igual o inferior a 0,42, lo que indica que los vinos con valores cercanos a 1 son poco frecuentes.

Residual Sugar:
El promedio es 2,5, valor maximo es 15.5 con desviación estándar 1,4. Significa que hay vinos con nivel de residual sugar 15.5, pero son muy poco frecuentes. 75% de los vinos tienen un nivel de Residual sugar igual o inferior a 2,6.

Chlorides:
Los vinos son muy similares en el nivel de cholides. Promedio: 0.08. Valores generalmente bajos que van de 0.01 a 0.06 con una desviación estándar muy pequeña, lo que indica poca variabilidad.

Free sulfur: 
Promedio: 15.8. Std 10,46. Valores entre 1 y 72. Variabilidad alta, 75% de los vinos con valores igual o inferior a 21. Unos pocos vinos tienen nivel de Free Sulfur entre 21 y 72.

Total Sulfur dioxide:
Aquí es donde se observa la mayor variabilidad entre los vinos, con valores que van desde 6 hasta 289.
La desviación estándar es 32,89, lo que indica que, aunque el promedio es 46, muchos vinos presentan niveles de total sulfur dioxide mucho más bajos o mucho más altos.

Density:
Los valores de densidad son muy similares entre los vinos; todos se encuentran entre 0.99 y 1.00.
No hay mucha variabilidad en esta característica.

pH:
Promedio 3,3 con std 0.15. No hay mucha diferencia en el nivel de pH entre los vinos.

Sulphates:
Promedio 0.65 con std 0.17. No hay mucha diferencia en el nivel de sulphates entre los vinos, pero ejemplo en el data de vino con nivel 2.0 que presente un valor atipico.

Alcohol:
Promedio: 10.4 con std 1.06. Valor minimo 8,4, maximo 14.9. 50% de los vinos tinen un % de alcohol igual o menos quw 10.2%

Quality:
Una variable que no miden y que es más subjetiva. Los vinos con peor calidad tienen una puntuación de 3, mientras que los mejores vinos alcanzan una calidad de 8. El promedio es 5.63 con una desviación estándar de 0.8 (baja). Solo el 25% de los vinos tiene una puntuación de 7 o superior.
