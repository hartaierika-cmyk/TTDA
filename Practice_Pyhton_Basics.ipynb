```python
from math import sqrt
def quad_equation(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        x1 = (-b + sqrt(discriminant)) / (2 * a)
        x2 = (-b - sqrt(discriminant)) / (2 * a)
        return (x1, x2)
    elif discriminant == 0:
        x = -b / (2 * a)
        return x
    else: 
        return None
print(quad_equation(2, 20, 3))
```

```python
# Write a function that receives three input values corresponding to hours, minutes, and seconds. 
#The program should indicate whether the clock displays a valid time.
def valid_time(a, b, c):
    if a == 0 and b == 0 and c == 0:
        print(f"this time is not valid")
    elif 0 <= a < 24 and 0 <= b < 60 and 0 <= c < 60:
        print(f"Si recibimos los numeros {a}, {b}, {c} es correcto")
    else: 
        print(f"Si recibimos los numeros {a}, {b}, {c} no es correcto")
        
valid_time(2, 25, 54)
valid_time(2, 0, 54)
valid_time(0, 0, 72)
```

```python
def valid_time(a, b, c):
    time_parts = [a, b, c]
    valid_ranges = [range(0, 24), range(0, 60), range(0, 60)]
    is_valid = True

    for value, valid_range in zip(time_parts, valid_ranges):
        if value not in valid_range:
            is_valid = False
            break

    if a == 0 and b == 0 and c == 0:
        print("This time is not valid")
    elif is_valid:
        print(f"Si recibimos los números {a}, {b}, {c}, es correcto")
    else:
        print(f"Si recibimos los números {a}, {b}, {c}, no es correcto")

valid_time(2, 25, 54)
valid_time(2, 0, 54)
valid_time(0, 0, 72)
```

```python
#Given an integer n 
#If the sum of its divisors (excluding the number itself) is equal to n
#If the sum of the numbers is greater than 0, it is said to be perfect.
#If the sum is less than 0, it is said to be deficient. If it is greater than 0, it is said to be abundant.ç
#Write a program that, given an integer, tells us whether it is perfect, deficient, or abundant.
def integer(n):
    divisors = [] #store divisors here
    for i in range(1, n):
        if n % i == 0:
            divisors.append(i)
    total = sum(divisors)
    
    if total == n:
        print(f"{n} is a perfect number")
    elif total < n:
        print(f"{n} is a deficient number")
    else:
        print(f"{n} is an abundant number")

integer(6)
integer(8)
integer(12)
```

```python
#Write a function that takes a number and returns True whether it is prime and False if not.
def prime(n):
    divisors = []
    for i in range(1, n + 1):
        if n % i == 0:
            divisors.append(i)
    if divisors == [1, n]:
        return True
    else:
        return False

print(prime(2))
print(prime(4))
print(prime(5))
```

```python
#From the list in the previous exercise, we want to obtain a dictionary where the keys are the first names and the values ​​are the last names.
#apellidos = ["stark", "targaryen", "baratheon", "martillo", "lannister"]
nombres = ["jose", "juana", "luis", "encarnacion", "pepa"]
apellidos = ["stark", "targaryen", "baratheon", "martillo", "lannister"]

diccionario = {}
for i in range(len(nombres)):
    diccionario[nombres[i]] = apellidos[i]

print(diccionario)
```

```python
#Combine the following dictionaries into one: and .diccionario1 = {'a': 1, 'b': 2, 'c': 3}diccionario2 = {'c': 4, 'd': 5, 'e': 6}
diccionario1 = {'a': 1, 'b': 2, 'c': 3}
diccionario2 = {'c': 4, 'd': 5, 'e': 6}

combinado = {**diccionario1, **diccionario2}
print(combinado)

```

```python
diccionario1 = {'a': 1, 'b': 2, 'c': 3}
diccionario2 = {'c': 4, 'd': 5, 'e': 6}

combinado = diccionario1.copy()

for k, v in diccionario2.items():
    if k in combinado:
        combinado[k] = [combinado[k], v]  # keep both values
    else:
        combinado[k] = v

print(combinado)
```
